```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/drive/MyDrive/datathon/dataset/train.csv')
```

```python
def findValue(col):
  plt.figure(figsize=(10,5))
  plt.title(col)
  sns.scatterplot(x=col, y='body fat_%', data=df)
  plt.show()
```

```python
for col in df.columns:
  findValue(col)
```
